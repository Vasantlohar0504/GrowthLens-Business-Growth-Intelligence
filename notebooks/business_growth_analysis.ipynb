```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("../data/raw/global_superstore_2016.xlsx")
df.head()

df.describe()

df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.show()

df.groupby("Region")["Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.show()
```
